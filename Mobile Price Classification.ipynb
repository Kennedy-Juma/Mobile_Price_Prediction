## Data Analysis

### Import the required libraries and loading the dataset

```python
#Import the libraries required for analysis
import pandas as pd
```

```python
#Loading the dataset
mobile_data=pd.read_csv('Dataset/mobile_price_classification.csv')
```

### Exploring and Understanding the dataset

```python
#Check the first 5 rows
mobile_data.head()
```

```python
#Check the last 5 rows 
mobile_data.tail()
```

```python
# Get number of rows and columns in the dataset
mobile_data.shape
```

The dataset has 2002 records/ rows and 22 columns.

```python
# Columns in the dataset
mobile_data.columns
```

```python
# Drop the id column
mobile_data=mobile_data.drop('id', axis=1)
```

The id column is dropped because it does not represent a feature of mobile phone. The remaining columns are mobile phone features.

```python
# Get a statistical overview of all numeric columns in the dataset
mobile_data.describe()
```

### Data Cleaning

#### Renaming columns

```python
#Rename columns
mobile_data=mobile_data.rename(columns={
                                        'blue':'bluetooth','fc':'fc_megapixel','m_dep':'m_depth','pc':'pc_megapixel'
                                        })
```

In this section, the stated columns are renamed to give the meaningful names

#### Checking for duplicate records and removing them

```python
#Check for duplicate values
duplicates=mobile_data.duplicated()
```

```python
duplicates.head()
```

```python
duplicates.sum()
```

There are 2 duplicate records in the dataset

```python
# Drop the duplicate records from the dataset
mobile_data=mobile_data.drop_duplicates()
```

```python
mobile_data.shape
```

The number of rows drops to 2000 from 2002 because we have dropped 2 rows that had duplicate records.

#### Checking and resolving null and missing values

```python
#Check for missing data using isna fuction
missing_data=mobile_data.isna().sum() #mobile_data.isnull().sum()
```

```python
missing_data
```

```python
#Drop the missing values
mobile_data=mobile_data.dropna()
```

```python
mobile_data.isnull().sum()
```

```python
mobile_data.dtypes
```

```python
#Export the cleaned data to a csv file
mobile_data.to_csv('Dataset/clean_mobile_data.csv', index=False)
```

### Exploratory Data Anaysis (EDA)

```python
#Import the required visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
#Read the cleaned dataset
mobile_data=pd.read_csv('Dataset/clean_mobile_data.csv')
```

```python
#Visualize the number of elements contained in each mobile phone price classes
mobile_data['price_range'].value_counts().plot(kind='bar')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
```

From the plot we can see that all the price ranges have same number of elements 

```python
#Distribution of bluetooth feature among the mobile phones
sns.displot(mobile_data['bluetooth'])
plt.show()
```

Half devices have bluetooth and half don't
